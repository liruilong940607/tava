```python
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F

from models import MLP
from utils import visualize_field
```

```python
model = MLP(
    input_dim=2, 
    net_depth=8, 
    net_width=16, 
    skip_layer=4, 
    output_dim=1,
)
```

```python
x, y, grad = visualize_field(model, bboxs=[0, 0, 1, 1], size=(500, 500))
canvas = grad.abs().sum(dim=-1).cpu().detach().numpy()
plt.imshow(canvas)
plt.show()
```

```python
x, y, grad = visualize_field(
    model, 
    loss_func=lambda x: F.mse_loss(x, model(torch.randn([1, 2])).detach()), 
    bboxs=[0, 0, 1, 1], 
    size=(500, 500)
)
canvas = grad.abs().sum(dim=-1).cpu().detach().numpy()
plt.imshow(canvas)
plt.show()
```
